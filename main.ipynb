```python
import numpy as np
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import pandas as pd
import glob
```

```python
#SAVING ORIGINAL SHAPE
shape_img_path = r'C:\Users\Polho\Downloads\archive\flower_images\flower_images_train\0001.png'
shape_img_path = Image.open(shape_img_path)
shape_img_path = np.array(shape_img_path)
shape_img = shape_img_path.shape
```

```python
#FOR EACH IMAGES IN DIR;
#open image
#tranform to array, after flatten and like register type "int64"
#add at list "train images" and convert for Pandas dataframe
#remove columns with null registers

train_images = []
for image in glob.glob(r'C:\Users\Polho\Downloads\archive\flower_images\flower_images_train\*.png'):
    images = Image.open(image)
    images = np.array(images).flatten().astype('int64')
    train_images.append(images)
df_flower = pd.DataFrame(train_images) 
df_flower = df_flower.dropna(axis=1)
```

```python
#From dataset, localizer row and convert to numpy.n-dimension array(numpy.ndarray)
#after make reshape of row with the tuple with original shape of all images

row = 4
img = df_flower.loc[row].to_numpy()
plt.imshow(img.reshape(shape_img))
plt.show()
```
